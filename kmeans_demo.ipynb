```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
```

```python
# Elbow method
error = []
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(input_data)
  error.append(kmeans.inertia_)
```

```python
plt.plot([i for i in range(1, 11)], error)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/house_rental_data.csv.txt")
```

```python
df
```

```python
x1 = df.sample(10, random_state = 10)
x2 = df.sample(10,  random_state = 9)
```

```python
x = df[['Sqft',	'Floor'	,'TotalFloor'	,'Bedroom',	'Living.Room',	'Bathroom']]
```

```python
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(5)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
knn.score(x_test, y_test)
```
