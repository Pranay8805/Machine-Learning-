```python
def minmaxscaling(x):
  min_val = np.min(x)
  max_val = np.max(x)
  return (x-min_val)/(max_val-min_val)
```

```python
df['scaled_salary'] = minmaxscaling(df['EstimatedSalary'])
```

```python
df['scaled_salary'].min(), df['scaled_salary'].max()
```

```python
def stdscaling(x):
  mean_val = np.mean(x)
  std_val = np.std(x)
  return (x-mean_val)/(std_val)
```

```python
df['scaled_std_salary'] = stdscaling(df['EstimatedSalary'])
```

```python
df['scaled_std_salary'].mean(), df['scaled_std_salary'].std()
```
